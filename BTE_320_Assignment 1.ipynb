```python
#Inputs
sales = input('Enter sales($): ')
sales = float(sales)
cogs = input('Enter Cost of Goods Sold(%): ')
cogs = float(cogs)
tax_rate = input('Enter Tax Rate(%): ')
tax_rate = float(tax_rate)

#Processes
COGS = float(sales) * (cogs / 100)
GrossProfit = (float(sales)) - (float(COGS))
TaxExpense = float(GrossProfit) * (tax_rate / 100)
NetIncome = float(GrossProfit) - float(TaxExpense)

#Outputs
print(f"Cost of Goods Sold: ${COGS}")
print(f"Gross Profit: ${GrossProfit}")
print(f"Tax Expense: ${TaxExpense}")
print(f"Net Income: ${NetIncome}")
```

```python
import random

def play_rock_paper_scissors():
    print("Welcome to Rock-Paper-Scissors!")
    print("Let's play against the computer!")

    # Main game loop
    play_again = "yes"
    while play_again.lower() == "yes":
        # Get user's choice
        user_choice = input("\nChoose rock, paper, or scissors: ").lower()

        # Validate user input
        while user_choice not in ['rock', 'paper', 'scissors']:
            print("Invalid choice! Please choose rock, paper, or scissors.")
            user_choice = input("Choose rock, paper, or scissors: ").lower()

        # Computer makes random choice
        computer_choice = random.choice(['rock', 'paper', 'scissors'])
        print(f"Computer chose: {computer_choice}")

        # Determine winner
        if user_choice == computer_choice:
            print("It's a tie!")
        elif (user_choice == 'rock' and computer_choice == 'scissors') or \
             (user_choice == 'paper' and computer_choice == 'rock') or \
             (user_choice == 'scissors' and computer_choice == 'paper'):
            print("You win!")
        else:
            print("Computer wins!")

        # Ask if user wants to play again
        play_again = input("\nDo you want to play again? (yes/no): ")

        # Validate play again input
        while play_again.lower() not in ['yes', 'no']:
            print("Please enter 'yes' or 'no'")
            play_again = input("Do you want to play again? (yes/no): ")

    print("\nThanks for playing!")

# Start the game
play_rock_paper_scissors()
```

```python
def getValues(n: int) -> list:
  '''
  Creates a list of weights.
  Input:
  - number of items in the list
  Output:
  - a list of weights
  '''
  Items = []

  for i in range(n):
    try:
      item = float(input(f'Enter weight values {i+1}: '))
    except Exception as e:
      print(e)
      item = float('nan')

    Items.append(item)
    print(Items)
  return Items

def calcAverage(alist: list) -> float:
  '''
  Input:
  - A list of weights
  Output:
  - Average value; excludes the first two items
  '''
  try:
    return sum(alist[2:]) / (len(alist[2:]))
  except ZeroDivisionError:
    return float('nan')

def main():
  num = int(input('Enter number of items: '))

  if num <= 0:
    raise ValueError('num must be positive')

  weights = getValues(num)
  avg = calcAverage(weights)
  print(f'Average value: {avg} lbs.')

main()
```

```python
def ROI(cost: float, revenue: float) -> float:
  assert cost > 0, "Cost must be positive"
  assert isinstance(cost, (int, float)), "Cost must be numerical"
  assert isinstance(revenue, (int, float)), "Revenue must be numerical"

  return (revenue - cost) / cost

if __name__ == '__main__':
  cost = float(input('Enter cost: '))
  revenue = float(input('Enter revenue: '))
  reInvest = ROI(cost,revenue)
  print(reInvest)
```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []

  def getName(self):
    return self.__name

  def getCourses(self):
    return self.__courses

  def setName(self, newname):
    self.__name = newname

  def __len__(self):
    return len(self.__courses)

  def enroll(self, course):
    if course in self.__courses:
      print('You have already enrolled in', course)
    else:
      self.__courses.append(course)

s = Student('Bob', 12345)
print(s.getName())
print(s.getCourses())
s.enroll('ABC123')
print(s.getCourses())
s.enroll('ABC123')
s.enroll('ABC456')
print(s.getCourses())
len(s)
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    x_diff_sq = (self.x - other.x) ** 2
    y_diff_sq = (self.y - other.y) ** 2

    return (x_diff_sq + y_diff_sq) ** 0.5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(1,0)
    self.p3 = Point(1,1)
    self.p4 = Point(0,1)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p4)

  def area(self):
    return self.width * self.height

sq = Rectangle()
sq.area()
```
